```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("signs.png")
```

```python
# copy to draw on
img1 = img.copy()
```

```python
img.shape
```

```python
# first row of image, should be 0-255
img[0]
```

## Drawing Bounding Boxes

```python
bbox = ((50, 50), (800, 800))
```

```python
# image is 
cv2.rectangle(img1, pt1=bbox[0], pt2=bbox[1], color=(0, 0, 255), thickness=20)
```

```python
plt.imshow(img1)
```

## Cropping Bounding Boxes

```python
tl, br = bbox
cropped = img[tl[0]:br[0], tl[1]:br[1], :]
```

```python
plt.imshow(cropped)
```
